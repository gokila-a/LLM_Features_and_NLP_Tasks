```python
from transformers import AutoTokenizer, AutoModelForQuestionAnswering
import torch

tokenizer = AutoTokenizer.from_pretrained("timpal0l/mdeberta-v3-base-squad2")
model = AutoModelForQuestionAnswering.from_pretrained("timpal0l/mdeberta-v3-base-squad2")

context = "i am from coimbatore,myself Gokila studying in KCE "
question = "what am i ?"

inputs = tokenizer(question, context, return_tensors="pt")
with torch.no_grad():
    outputs = model(**inputs)

answer_start_index = outputs.start_logits.argmax()
answer_end_index = outputs.end_logits.argmax() + 1

answer = tokenizer.convert_tokens_to_string(tokenizer.convert_ids_to_tokens(inputs["input_ids"][0][answer_start_index:answer_end_index]))
print(f"Question: {question}")
print(f"Context: {context}")
print(f"Answer: {answer}")
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

tokenizer = AutoTokenizer.from_pretrained("Helsinki-NLP/opus-mt-en-fr")
model = AutoModelForSeq2SeqLM.from_pretrained("Helsinki-NLP/opus-mt-en-fr")

text_to_translate = "I am gokila from Coimbatore."

inputs = tokenizer(text_to_translate, return_tensors="pt")
outputs = model.generate(**inputs)
translated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)

print(f"Original text: {text_to_translate}")
print(f"Translated text (French): {translated_text}")
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
import torch

tokenizer = AutoTokenizer.from_pretrained("facebook/bart-large-cnn")
model = AutoModelForSeq2SeqLM.from_pretrained("facebook/bart-large-cnn")

text_to_summarize = """Your task is to generate JSON. You are provided with a schema and an incomplete JSON instance. You must complete the JSON instance according to the schema. Do not output anything else. If a field's type is array, and the incomplete JSON does not already contain an array for this field, you can generate an empty array if there are no items to add. If a field's type is string, and the incomplete JSON does not already contain a string for this field, you can generate an empty string if there are no characters to add. Do not generate null values for required fields. Your entire response must be a valid JSON instance. The example responses are for illustrative purposes only, and may not fully conform to the schema as they are not used to generate JSON. However, your response must be valid JSON according to the schema.Your task is to generate JSON. You are provided with a schema and an incomplete JSON instance. You must complete the JSON instance according to the schema. Do not output anything else. If a field's type is array, and the incomplete JSON does not already contain an array for this field, you can generate an empty array if there are no items to add. If a field's type is string, and the incomplete JSON does not already contain a string for this field, you can generate an empty string if there are no characters to add. Do not generate null values for required fields. Your entire response must be a valid JSON instance. The example responses are for illustrative purposes only, and may not fully conform to the schema as they are not used to generate JSON. However, your response must be valid JSON according to the schema."""

# Encode the text
inputs = tokenizer(text_to_summarize, return_tensors="pt", max_length=1024, truncation=True)

# Generate summary
summary_ids = model.generate(
    inputs["input_ids"],
    num_beams=4,
    min_length=25,
    max_length=50,
    early_stopping=True,
    do_sample=False
)

summary_text = tokenizer.decode(summary_ids[0], skip_special_tokens=True)

print(f"Original text: {text_to_summarize[:200]}...")
print(f"Summary: {summary_text}")
```
